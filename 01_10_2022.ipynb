````python
img=cv2.imread('/content/photo_1536937275673_8a12b3812e8e.jpeg',0)
plt.imshow(img)
````

````python
#convert BGR to RGB
import cv2
from cv2 import imread as imread
from matplotlib import pyplot as plt
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
````

````python
#draw border
img = io.imread('/content/photo_1536937275673_8a12b3812e8e.jpeg',0)
b=cv2.copyMakeBorder(img,40,40,40,40,cv2.BORDER_ISOLATED)
plt.imshow(b)
````

````python
#saving image
cv2.imwrite('output.jpg',b)

````

````python
#image rotation
rotate=cv2.rotate(img_rgb,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotate)
````

````python
#finding shape
img_rgb.shape
(253,247,3)
````

````python
#resize image
img_resize=cv2.resize(img_rgb,(400,400))
plt.imshow(img_resize)
````

````python
#finding shape
img_resize.shape
````

````python
#crop image
crop=img_resize[150:320:,100:350]
plt.imshow(crop)
````

````python
#blur image
blur=cv2.GaussianBlur(img_resize,(13,0),cv2.BORDER_DEFAULT)
plt.imshow(blur)
````
